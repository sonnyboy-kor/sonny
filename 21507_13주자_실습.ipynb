```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('20210528', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns = ['A', 'B', 'C', 'D'])
```

```python
df.plot()
df = df.cumsum()
plt.show()
```
