```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('20210528', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
plt.show()
```
